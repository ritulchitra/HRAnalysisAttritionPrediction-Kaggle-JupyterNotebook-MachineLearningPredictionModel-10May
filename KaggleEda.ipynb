# **Age relation with Attrition**

```python
plt.figure(figsize=(12, 6))

# Create age groups
age_bins = [18, 25, 30, 35, 40, 45, 50, 55, 60, 65]
age_labels = ['18-24', '25-29', '30-34', '35-39', '40-44', '45-49', '50-54', '55-59', '60+']
df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

# Calculate percentages (with observed=True to prevent warning)
attrition_counts = df.groupby(['AgeGroup', 'Attrition'], observed=True).size().unstack()
attrition_pct = attrition_counts.div(attrition_counts.sum(axis=1), axis=0) * 100

# Set style - using 'seaborn-v0_8' which is the modern equivalent
plt.style.use('seaborn-v0_8')

# Plot the lines
sns.lineplot(data=attrition_pct, x=attrition_pct.index, y='No', 
             color='#2ecc71', label='Still Working (No)', linewidth=3, marker='o', markersize=8)
sns.lineplot(data=attrition_pct, x=attrition_pct.index, y='Yes', 
             color='#e74c3c', label='Left Company (Yes)', linewidth=3, marker='o', markersize=8)

# Customizations
plt.title('Employee Attrition Percentage by Age Group\n(What % of each age group leaves/stays?)', 
          fontsize=16, pad=20, fontweight='bold')
plt.xlabel('Age Group', fontsize=13, labelpad=10)
plt.ylabel('Percentage of Employees (%)', fontsize=13, labelpad=10)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.grid(True, linestyle='--', alpha=0.4)

# Add data labels
for line in plt.gca().get_lines():
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        plt.text(x, y+1, f'{y:.1f}%', color=line.get_color(), 
                 ha='center', va='bottom', fontsize=10, fontweight='bold')

# Highlight peak attrition age group
max_attrition_age = attrition_pct['Yes'].idxmax()
max_pct = attrition_pct['Yes'].max()
plt.axvline(x=max_attrition_age, color='#e74c3c', linestyle=':', alpha=0.6)
plt.text(max_attrition_age, max_pct+5, f'Peak Attrition: {max_pct:.1f}%', 
         color='#e74c3c', ha='center', fontsize=11, bbox=dict(facecolor='white', alpha=0.8))

plt.legend(title='Attrition Status', fontsize=11, title_fontsize=12, 
           loc='upper right', framealpha=1)
plt.tight_layout()
plt.show()
```

# **Educational Background**

```python

# Fix common typos in Education values
df['Education'] = df['Education'].replace({
    'Colledge': 'College',
    'Below Colledge': 'Below College'
})

# Create cross-tabulation
education_attrition = pd.crosstab(df['Education'], df['Attrition'])

# Calculate percentages
education_attrition['Percentage_Yes'] = (
    education_attrition['Yes'] / (education_attrition['No'] + education_attrition['Yes']) * 100
)

# Order education levels
education_order = ['Below College', 'College', 'Bachelor', 'Master', 'Doctor']
education_attrition = education_attrition.reindex(education_order)

# Reset index for plotting
education_attrition = education_attrition.reset_index()
education_attrition = education_attrition.rename(columns={'index': 'Education'})

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(
    x='Education', 
    y='Percentage_Yes', 
    hue='Education', 
    data=education_attrition, 
    palette='viridis', 
    legend=False
)

# Customize plot
plt.title('Percentage of Attrition by Education Level', fontsize=16)
plt.xlabel('Education Level', fontsize=12)
plt.ylabel('Percentage with Attrition (Yes)', fontsize=12)
plt.ylim(0, 50)

# Add labels
for index, value in enumerate(education_attrition['Percentage_Yes']):
    if pd.notna(value):
        plt.text(index, value + 1, f'{value:.1f}%', ha='center', va='bottom')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create cross-tabulation for EducationField vs Attrition
edu_field_attrition = pd.crosstab(df['EducationField'], df['Attrition'])

# Calculate percentage of 'Yes' attrition
edu_field_attrition['Percentage_Yes'] = (
    edu_field_attrition['Yes'] / (edu_field_attrition['No'] + edu_field_attrition['Yes']) * 100
)

# Optional: Sort by percentage (or keep as is)
edu_field_attrition = edu_field_attrition.sort_values('Percentage_Yes', ascending=False)

# Reset index for Seaborn
edu_field_attrition = edu_field_attrition.reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(
    x='EducationField',
    y='Percentage_Yes',
    hue='EducationField',
    data=edu_field_attrition,
    palette='magma',
    legend=False
)

# Customize plot
plt.title('Percentage of Attrition by Education Field', fontsize=16)
plt.xlabel('Education Field', fontsize=12)
plt.ylabel('Percentage with Attrition (Yes)', fontsize=12)
plt.ylim(0, 50)

# Add percentage labels
for index, value in enumerate(edu_field_attrition['Percentage_Yes']):
    if pd.notna(value):
        plt.text(index, value + 1, f'{value:.1f}%', ha='center', va='bottom')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# **Satisfaction Level**

```python
# Define the columns to analyze
satisfaction_cols = [
    'EnvironmentSatisfaction',
    'JobInvolvement',
    'JobSatisfaction',
    'RelationshipSatisfaction'
]

# Define the satisfaction levels
levels = ['Low', 'Medium', 'High', 'Very High']

# Initialize a dictionary to store percentage values
percentage_data = {col: [] for col in satisfaction_cols}

# Calculate percentage of 'Attrition' == 'Yes' for each level in each column
for col in satisfaction_cols:
    for level in levels:
        total = len(df[df[col] == level])
        if total == 0:
            percentage = 0
        else:
            yes_count = len(df[(df[col] == level) & (df['Attrition'] == 'Yes')])
            percentage = (yes_count / total) * 100
        percentage_data[col].append(percentage)

# Plotting the line graph
plt.figure(figsize=(12, 6))

for col in satisfaction_cols:
    plt.plot(levels, percentage_data[col], marker='o', label=col)

plt.title('Attrition % by Satisfaction Levels')
plt.xlabel('Satisfaction Level')
plt.ylabel('Attrition Percentage (%)')
plt.ylim(0, 50)  # Limit Y-axis from 0 to 50%
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# **Job Role & Gender affecting Attrition**

```python
# Filter the dataframe to get only rows where Attrition is 'Yes'
attrition_yes = df[df['Attrition'] == 'Yes']

# Count the number of men and women who have Attrition = Yes
male_attrition_yes = attrition_yes[attrition_yes['Gender'] == 'Male'].shape[0]
female_attrition_yes = attrition_yes[attrition_yes['Gender'] == 'Female'].shape[0]

# Count the total number of men and women in the entire dataset
total_men = df[df['Gender'] == 'Male'].shape[0]
total_women = df[df['Gender'] == 'Female'].shape[0]

# Calculate percentage of attrition among men and women
male_attrition_pct = (male_attrition_yes / total_men) * 100
female_attrition_pct = (female_attrition_yes / total_women) * 100

# Prepare data for the pie chart
labels = ['Men Attrition %', 'Women Attrition %']
sizes = [male_attrition_pct, female_attrition_pct]
colors = ['skyblue', 'lightcoral']

# Plotting the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Attrition Percentage by Gender')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
# Filter the dataframe for Attrition = 'Yes'
attrition_yes = df[df['Attrition'] == 'Yes']

# Get all unique job roles from the JobRole column
job_roles = df['JobRole'].unique()

# Initialize a list to store the percentage attrition for each job role
attrition_percentages = []

# Loop through each job role and calculate the percentage of attrition
for role in job_roles:
    total_in_role = df[df['JobRole'] == role].shape[0]
    attrition_in_role = attrition_yes[attrition_yes['JobRole'] == role].shape[0]
    
    # Calculate percentage and add to list
    percentage = (attrition_in_role / total_in_role) * 100
    attrition_percentages.append(percentage)

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(attrition_percentages, labels=job_roles, autopct='%1.1f%%', startangle=140)
plt.title('Attrition Percentage by Job Role')
plt.axis('equal')  # Ensures pie chart is a circle
plt.show()
```

# **Attrition affected by Money**

```python
# Filter the dataset for employees who left the company
attrition_yes = df[df['Attrition'] == 'Yes']

# Create income bins for better visualization (e.g., every 20 bins)
income_bins = pd.cut(attrition_yes['MonthlyIncome'], bins=20)

# Count number of attrition cases per income bin, using observed=True to avoid warning
income_trend = attrition_yes.groupby(income_bins, observed=True).size()

# Get bin centers for smooth x-axis
bin_centers = [interval.mid for interval in income_trend.index]

# Plotting the smooth line chart
plt.figure(figsize=(10, 6))
sns.lineplot(x=bin_centers, y=income_trend.values, marker='o')
plt.xlabel('Monthly Income')
plt.ylabel('Number of Attrition = Yes')
plt.title('Trend of Attrition Count vs Monthly Income')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Convert Attrition to numeric (1 for 'Yes', 0 for 'No')
df['AttritionNumeric'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# Calculate rolling percentages (smoother than bins)
window_size = 50  # Adjust this for more/less smoothing
df_sorted = df.sort_values('DailyRate')
rolling_pct = df_sorted['AttritionNumeric'].rolling(window=window_size).mean() * 100

# Plot with clean lines
plt.plot(df_sorted['DailyRate'], rolling_pct, 
         color='#e74c3c', label='Left Company (Yes)', linewidth=2)
plt.plot(df_sorted['DailyRate'], 100-rolling_pct, 
         color='#2ecc71', label='Still Working (No)', linewidth=2)

# Customizations
plt.title('Attrition Percentage by Daily Rate', fontsize=16, pad=20)
plt.xlabel('Daily Rate ($)', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.legend(fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Highlight key areas
max_attrition = rolling_pct.max()
plt.axhline(y=max_attrition, color='#e74c3c', linestyle=':', alpha=0.5)
plt.text(df['DailyRate'].max()*0.8, max_attrition+2, 
         f'Max attrition: {max_attrition:.1f}%', color='#e74c3c')

plt.tight_layout()
plt.show()
```
